```python
import pandas as pd
import matplotlib as plt

drug_pricing = pd.read_csv("Historical Drug Pricing Data.csv")

drug_pricing = drug_pricing.sort_values(by='NDC2', ascending = True)

# collapsing data to a one dimensional space, see if there are any correlation between features

# drug_pricing.info()

# drug_pricing.nunique()
#drug_pricing
```

```python
analysis_summary = drug_pricing.drop_duplicates(subset=['NDC2'], keep='first')
analysis_summary = analysis_summary.iloc[:,:5]
#print(analysis_summary)
```

```python
# average WAC
average_wac = drug_pricing.groupby('NDC2')['WAC per Unit'].mean()
#print(average_wac)
analysis_summary = pd.merge(analysis_summary, average_wac, on=['NDC2'],how='left')
#analysis_summary
```

```python
# average AWP
average_awp = drug_pricing.groupby('NDC2')['AWP per Unit'].mean()
analysis_summary = pd.merge(analysis_summary, average_awp, on=['NDC2'],how='left')
#analysis_summary
```

```python
drug_pricing['Date'] = pd.to_datetime(drug_pricing['year'].astype(str) + '-' + drug_pricing['month'].astype(str), format='%Y-%m')
drug_pricing.sort_values(by='Date', inplace=True)
drug_pricing['WAC Change'] = drug_pricing.groupby('NDC2')['WAC per Unit'].diff()
WAC_total_change = drug_pricing.groupby('NDC2')['WAC Change'].sum()
analysis_summary = pd.merge(analysis_summary, WAC_total_change, on=['NDC2'],how='left')
```

```python
drug_pricing['AWP Change'] = drug_pricing.groupby('NDC2')['AWP per Unit'].diff()
AWP_total_change = drug_pricing.groupby('NDC2')['AWP Change'].sum()

analysis_summary = pd.merge(analysis_summary, AWP_total_change, on=['NDC2'],how='left')
analysis_summary
```
