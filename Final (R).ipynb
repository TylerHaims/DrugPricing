```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
drugs = pd.read_csv('Historical Drug Pricing Data.csv')
```

```python
drugs[drugs['HCPCS_Code'].str[0].str.islower()]
drugs['HCPCS_Code'] = drugs['HCPCS_Code'].str.upper()
drugs['Manufacturer'] = drugs['Manufacturer'].str.upper()
```

```python
drugs['Code'] = drugs['HCPCS_Code'].str[0]
```

```python
drugs = drugs[drugs['Code'] != 'N']
```

```python
len(drugs['Manufacturer'].unique())
```

```python
group = drugs.groupby('HCPCS_Code')['Manufacturer'].nunique()


mask = group > 1

HCPCS = drugs.copy()

# Initialize the 'generic' column with zeros
HCPCS['generic'] = 0

# Set the index of the DataFrame to 'HCPCS_Code'
HCPCS.set_index('HCPCS_Code', inplace=True)

# Set the values in the 'generic' column based on the mask using .loc
HCPCS.loc[mask.index, 'generic'] = mask.astype(int)

# Reset the index to bring 'HCPCS_Code' back as a column
HCPCS.reset_index(inplace=True)
```

```python
HCPCS
```

```python
HCPCS.drop_duplicates(subset='HCPCS_Code', keep='first', inplace=True)
```

```python
HCPCS
```

```python
HCPCS.drop(columns=['NDC2', 'Billing Units per Package', 'Medicare Payment Limit', 'Average Sales Price', 'APC Payment Limit', 'WAC per Package', 'WAC per Unit', 'WAC Effective Date', "AWP per Package", 'AWP per Unit', 'AWP Effective Date', 'month', 'year', 'dif', 'present'], inplace=True)
```

```python
HCPCS.drop(columns=['HCPCS Description', 'Drug Name', 'Manufacturer'], inplace=True)
```

```python
HCPCS['HCPCS_Code'].to_csv('output.txt', index=False, sep='\t')
```
