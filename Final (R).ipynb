```python
group = drugs.groupby('HCPCS_Code')['Manufacturer'].nunique()


mask = group > 1

HCPCS = drugs.copy()

# Initialize the 'generic' column with zeros
HCPCS['generic'] = 0

# Set the index of the DataFrame to 'HCPCS_Code'
HCPCS.set_index('HCPCS_Code', inplace=True)

# Set the values in the 'generic' column based on the mask using .loc
HCPCS.loc[mask.index, 'generic'] = mask.astype(int)

# Reset the index to bring 'HCPCS_Code' back as a column
HCPCS.reset_index(inplace=True)
```

```python
HCPCS
```

```python
HCPCS.drop_duplicates(subset='HCPCS_Code', keep='first', inplace=True)
```

```python
HCPCS
```

```python
HCPCS.drop(columns=['NDC2', 'Billing Units per Package', 'Medicare Payment Limit', 'Average Sales Price', 'APC Payment Limit', 'WAC per Package', 'WAC per Unit', 'WAC Effective Date', "AWP per Package", 'AWP per Unit', 'AWP Effective Date', 'month', 'year'], inplace=True)
```

```python
HCPCS.drop(columns=['HCPCS Description', 'Drug Name', 'Manufacturer'], inplace=True)
```

```python
drug_pricing = pd.read_csv("Historical Drug Pricing Data.csv")

drug_pricing = drug_pricing.sort_values(by='NDC2', ascending = True)

# analysis summary used to give an overall picture of the data. it keeps the hcpc code, has unique ndcs, and will be used to store average price, etc
analysis_summary = drug_pricing.drop_duplicates(subset=['NDC2'], keep='first')
analysis_summary = analysis_summary.iloc[:,:5]
# average WAC
average_wac = drug_pricing.groupby('NDC2')['WAC per Unit'].mean()
analysis_summary = pd.merge(analysis_summary, average_wac, on=['NDC2'],how='left')

chosen_ndc = analysis_summary.groupby('HCPCS_Code')['WAC per Unit'].max()
# chosen_ndc looks at the ndcs with the HCPCS and chooses the highest average price
# it doesn't give you which ndc corresponds to that average price
```
